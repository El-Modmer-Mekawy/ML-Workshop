# Exercise 6: Vision-Based Control

## Objective
Use YOLO detections to generate control commands.

If object is:
- Left side → Print 'Turn Left'
- Right side → Print 'Turn Right'
- Center → Print 'Move Forward'

## Step 1: Import Libraries

```python
from ultralytics import YOLO
import cv2
import numpy as np
```

## Step 2: Load Trained Model

```python
model = YOLO('runs/detect/train/weights/best.pt')
```

## Step 3: Vision-Based Command Execution

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    height, width, _ = frame.shape

    results = model(frame)
    annotated_frame = results[0].plot()

    if len(results[0].boxes) > 0:
        box = results[0].boxes[0]
        x_center = int(box.xywh[0][0])

        if x_center < width / 3:
            command = 'Turn Left'
        elif x_center > 2 * width / 3:
            command = 'Turn Right'
        else:
            command = 'Move Forward'

        print(command)

    cv2.imshow('Vision Control', annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
